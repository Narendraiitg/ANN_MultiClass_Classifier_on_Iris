```python
# Importing the various libraries
import seaborn as sns
import numpy as np
import keras
from keras.models import Sequential
from keras.optimizers import Adam
from keras.utils.vis_utils import model_to_dot
from keras.utils import plot_model
import pydot
import graphviz
from IPython.display import SVG
```

```python
dataset= pd.read_csv("iris.data")
dataset
```

# Data Preprocessing

```python
# Seperating the dependent and independent variables
X= data.iloc[:,0:4].values
y= data.iloc[:, 4].values
```

```python
# Encoding categorial data 
from sklearn.preprocessing import LabelEncoder
LE= LabelEncoder()
y1= LE.fit_transform(y)
Y= pd.get_dummies(y1).values # converting the different values of y1 in the form of matrix
```

```python
y1
```

```python
Y
```

In above [1,0,0] represent Iris-setosa, [0,1,0] represent Iris-versicolor and [0,0,1] represent Iris-virginica

```python
# Spiliting the data set into training and testing 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,Y, test_size=0.3, random_state=0)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Adding the input and first hidden layer 
model= Sequential()

#The first hidden layer has 16 nodes and uses the relu activation function.
model.add(Dense(4, input_shape=(4,), activation='relu'))

# Adding the output layer
model.add(Dense(3, activation='softmax')) # here we used softmax activation in output which will decide the output on the basis of max prob of all 3 catg of o/p
```

```python
model.compile(optimizer="Adam", loss='categorical_crossentropy', metrics= ['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=10, epochs=100, verbose=0)
```

```python
# make predictions with the model
y_pred = model.predict(X_test)
y_test_class= np.argmax(y_test,axis=1)
y_pred_class= np.argmax(y_pred,axis=1)
```

```python
# Making the confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test_class, y_pred_class))
print(confusion_matrix(y_test_class, y_pred_class))
```
